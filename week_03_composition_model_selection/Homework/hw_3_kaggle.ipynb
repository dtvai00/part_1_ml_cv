Построим корреляционную матрицу для поиска линейно связанных между собой признаков

```python
def plot_correlation_matrix(df, threshold=0.8):
    
    corr_matrix = df.corr()
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, linewidths=0.5)
    plt.title('Матрица корреляций')
    plt.show()
    
    strong_corr = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                strong_corr.append((
                    corr_matrix.columns[i],
                    corr_matrix.columns[j],
                    corr_matrix.iloc[i, j]
                ))
    
    print(f"\nСильно коррелированные пары (|corr| > {threshold}):")
    for f1, f2, corr in strong_corr:
        print(f"  {f1} - {f2}: {corr:.3f}")
    
    return corr_matrix, strong_corr

corr_matrix, strong_pairs = plot_correlation_matrix(data[num_cols+[target_col]], threshold=0.85)
```

```python
# Кодируем выбранные категориальные признаки в one-hot числовые столбцы.
enc = OneHotEncoder(drop='first',sparse_output=False) 

X_train_encoded = enc.fit_transform(X_train[cat_to_keep])
X_valid_encoded = enc.transform(X_valid[cat_to_keep])
X_test_encoded = enc.transform(X_test[cat_to_keep])

feature_names = enc.get_feature_names_out(cat_to_keep)
one_hot_train = pd.DataFrame(X_train_encoded, columns=feature_names)
one_hot_valid = pd.DataFrame(X_valid_encoded, columns=feature_names)
one_hot_test = pd.DataFrame(X_test_encoded, columns=feature_names)

X_train_enc = pd.merge(X_train[num_cols].reset_index(drop=True), one_hot_train, left_index = True, right_index=True)
X_valid_enc = pd.merge(X_valid[num_cols].reset_index(drop=True), one_hot_valid, left_index = True, right_index=True)
X_test_enc = pd.merge(X_test[num_cols].reset_index(drop=True), one_hot_test, left_index = True, right_index=True)
```

```python
# Стандартизация
scaler = StandardScaler()
X_train_stand = scaler.fit_transform(X_train_enc)
X_valid_stand = scaler.transform(X_valid_enc)
X_test_stand = scaler.transform(X_test_enc)

# Обучение модели
C_l = [100, 10, 20, 30, 40, 50, 5, 7, 0.1, 0.01, 0.001] # Перебор значений гиперпараметра C от 0.001 до 100
clf_reg = LogisticRegressionCV(Cs = C_l, scoring = 'roc_auc',refit=True, random_state=42)

clf_reg.fit(X_train_stand, y_train)

y_tp = clf_reg.predict_proba(X_train_stand)[:, 1]
y_vp = clf_reg.predict_proba(X_valid_stand)[:, 1]

print(
    f'Train: {roc_auc_score(y_train, y_tp)}\n'
    f'valid: {roc_auc_score(y_valid, y_vp)}\n'
)
```

```python
import optuna
from catboost import CatBoostClassifier, Pool, cv
```

### Подбор параметров модели с помощью Optuna

```python
def objective(trial):
    
    params = {
        'iterations': trial.suggest_int('iterations', 100, 500),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.5, log=True),
        'depth': trial.suggest_int('depth', 5, 10),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1, 10),
        'random_strength': trial.suggest_float('random_strength', 0.1, 10),
        'bagging_temperature': trial.suggest_float('bagging_temperature', 0, 1),
        'border_count': trial.suggest_int('border_count', 32, 255),
        'grow_policy': trial.suggest_categorical('grow_policy', ['SymmetricTree', 'Depthwise']),
        'loss_function': 'Logloss',
        'eval_metric': 'AUC',
        'verbose': False,
        'random_seed': 42,
        'task_type': 'GPU',
        'auto_class_weights': trial.suggest_categorical('auto_class_weights', ['Balanced', 'SqrtBalanced'])
    }
    

    model = CatBoostClassifier(**params)
    
    model.fit(
        X_train, y_train,
        eval_set=(X_valid, y_valid),
        early_stopping_rounds=50,
        verbose=False,
        cat_features=cat_to_keep
    )
    
    y_val_pred = model.predict_proba(X_valid)[:, 1]
    val_score = roc_auc_score(y_valid, y_val_pred)
    
    trial.set_user_attr('model', model)
    
    return val_score


study = optuna.create_study(
    direction='maximize',
    study_name='catboost_hyperparams',
    storage=None, 
    load_if_exists=False,
    pruner=optuna.pruners.MedianPruner(n_warmup_steps=10)
)


study.optimize(objective, n_trials=50, timeout=3600)

print(f'Best validation AUC: {study.best_value:.4f}')
print(f'Best params: {study.best_params}')
```

```python
params = study.best_params
params['verbose'] = False
params['random_state'] = 42
params['task_type'] = 'GPU'
params['cat_features'] = cat_to_keep

clf_catboost = CatBoostClassifier(**params)
clf_catboost.fit(X_train, y_train)
y_tp_c = clf_catboost.predict_proba(X_train)[:, 1]
y_vp_c = clf_catboost.predict_proba(X_valid)[:, 1]

print(
    f'Train: {roc_auc_score(y_train, y_tp_c)}\n'
    f'valid: {roc_auc_score(y_valid, y_vp_c)}\n'
)

# Предсказание вероятностей
# probabilities = clf_catboost.predict_proba(X_test)[:, 1]
```

```python
y_train_pred = y_tp_c
y_valid_pred = y_vp_c
train_auc = roc_auc_score(y_train, y_train_pred)
test_auc = roc_auc_score(y_valid, y_valid_pred)

plt.figure(figsize=(10,7))
plt.plot(*roc_curve(y_train, y_train_pred)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_valid, y_valid_pred)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.title(f'CatBoost: max_depth = {params['depth']}, n_estimators = {params['iterations']}, learning_rate = {round(params['learning_rate'],4)}')
plt.show()
```
